### <font color='#8B0000'>Q3 Use the country-level data set to create a histogram to:</font>
- Show the distribution of values for the number of hospitalizations per 1000 people.  
- Draw a red vertical line that shows the median value on the histogram.

```python
data_dropna = data.dropna(subset=['hosp'])
hosp = data_dropna.groupby('id').mean().hosp.reset_index()
hosp.head()
```

```python
def hosp_per_1000(id):
    return hosp[hosp.id == id]['hosp'].values[0] / data_dropna[data_dropna.id == id].population.unique()[0] * 1000
```

```python
hosp['hosp_per_1000'] = hosp['id'].apply(hosp_per_1000)
hosp.head()
```

```python
f = hosp.hosp_per_1000.plot(kind='hist', title='Histogram of number of hospitalizations per 1000 people')
f.axvline(x=hosp.hosp_per_1000.median(), color='r')
```

```python
hosp[hosp.hosp_per_1000 > 0.6]
```

```python
f = hosp.hosp_per_1000.plot(kind='hist', range=[0, 0.6], bins=20, title='Histogram of number of hospitalizations per 1000 people')
f.axvline(x=hosp.hosp_per_1000.median(), color='r')
```

As `hosp` is the number of hospitalized patients on date, and here I would like to find the distribution of values for the number of hospitalizations per 1000 people across countries, I just drop the null values, group `hosp` by countries, and take the means of `hosp` over time. It might be reasonable to keep integer values for the the number of hospitalizations per 1000 people, but as the values are small here, I just keep the floating numbers. Then I create a histogram, which has an outlier. Since this outlier is the only value that is bigger than 0.6, I zoom in to create another histogram without the outlier, The red vertical line shows the median value on the histogram.

From this histogram, we can see the values for the number of hospitalizations per 1000 people are mostly concentrated in the range from 0 to 0.3. In general, number of hospitalizations per 1000 people seems to be relatively low across the countries where the `hosp` data is available.

### <font color='#8B0000'> Q4 Create a visualization that shows the  number of tests per day in the United States and indicate the median value on your visualization.</font>

```python
usa, src = covid19('USA', verbose=False)
usa_tests = usa.groupby('date').sum().tests.reset_index()
usa_tests.head()
```

```python
usa_tests.sort_values('date').tail(10)
```

```python
usa_tests.drop(usa_tests.index[394: ],inplace=True)
usa_tests
```

```python
def tests_per_day(date):
    if date == pd.Timestamp('2020-01-22'):
        return usa_tests[usa_tests.date == date].tests.values[0]
    else:
        yesterday = date - pd.to_timedelta(24, unit='h')
        return usa_tests[usa_tests.date == date].tests.values[0] - usa_tests[
            usa_tests.date == yesterday].tests.values[0]
```

```python
usa_tests['test_per_day'] = usa_tests['date'].apply(tests_per_day)
usa_tests['test_per_day']
```

```python
plt.scatter(x='date', y='test_per_day', data=usa_tests)
plt.axhline(y=usa_tests['test_per_day'].median(), color='g')
plt.text(pd.Timestamp('2020-02-15'), 0.9e6, 'median='+str(usa_tests['test_per_day'].median()))
plt.show()
```

Here I dropped the rows starting from index 394 to the end from the raw data, as the cumulative number of tests for these days are 0, which is strange, and indicates that the data for these days have not been updated (Thus, if the notebook is run later, these values might be updated. However, to ensure that I have valid points here, I just drop the invalid values for the time being.) Then I define a function `test_per_day` to split the cumulative number of tests into number of tests per day, and create a scatter plot to visualize it. The median value is indicated by the green line.

From the plot, we can see the number of tests per day in the United States increases until November 2020, and then remains a relatively stable state where the number of tests per day is above the median value.
